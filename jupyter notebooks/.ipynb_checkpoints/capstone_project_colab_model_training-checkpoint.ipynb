**DATA PREPROCESSING**

We will be using Image Generator to load the data from folder using the DirectoryIterator method,

this ensures efficient data transmission and feeding of data for training and testing of Deep Learning Model.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
img_size = [255,255]
train_path = '/content/data/train'
test_path = '/content/data/test'

train_gen = ImageDataGenerator(rescale=1./255,
                               shear_range=0.2,
                               zoom_range=0.2,
                               horizontal_flip=True)

test_gen = ImageDataGenerator(rescale=1./255)


train_dataset = train_gen.flow_from_directory(train_path,
                      target_size = img_size,
                      batch_size = 16,
                      class_mode = 'categorical')
test_dataset = test_gen.flow_from_directory(test_path,
                        target_size = img_size,
                        batch_size = 16,
                        class_mode = 'categorical')
```

```python
print(f'Classes and their target variable in Train Dataset for classification is given by: \n {train_dataset.class_indices}')
```

```python
print(f'Classes and their target variable in Test Dataset for classification is given by: \n {test_dataset.class_indices}')
```

```python
print(f'Number of batches in Train Dataset is : {len(train_dataset)}')
print(f'Number of batches in Test Dataset is : {len(test_dataset)}')
```

## Section 1:

**MODEL INITIALISING**

Making a tensorflow.keras model with help of Conv2D and Maxpool layers.

```python
from tensorflow.keras.models import Model, load_model, Sequential
from tensorflow.keras.callbacks import ModelCheckpoint, CSVLogger
from tensorflow import argmax
from tensorflow import keras
from keras.utils import plot_model
import numpy as np
```

```python
img_size = [255,255]
```

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=img_size + [3]))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(64,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(128,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(128,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(512,activation='relu'))
model.add(keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
checkpoint_path2 = '/content/drive/MyDrive/capstone-project/training_log_custom/model_checkpoint/'
csv_log_path = '/content/drive/MyDrive/capstone-project/training_log_custom/logs/loss_acc_log2.csv'
#checkpoint_path = '/content/drive/MyDrive/capstone-project/training_log/model_checkpoints'

checkpoint_2 = ModelCheckpoint(filepath = checkpoint_path2,
                             save_weights_only=True,
                             save_best_only=False)
csv_logger_2 = CSVLogger(csv_log_path)

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
r2 = model.fit(train_dataset, validation_data = test_dataset,
                epochs=20,

                callbacks = [checkpoint_2,early_stopping, csv_logger_2])
```

# Metrics:

We will be employing different metrics to understand and analyse the efficiency and accuracy of our Model.

```python
def metrics_report(true_label, pred_label):
    accuracy = accuracy_score(true_label,pred_label)
    precision = precision_score(true_label,pred_label, average='macro')  # using macro because our data is balanced with all labels
    recall = recall_score(true_label, pred_label, average='macro')
    f1 = f1_score(true_label, pred_label, average='macro')
    

    metrics_dict = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
    }

    return metrics_dict
```

## Model Testing

Will feed image of forest and predict the class:

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np

def predictImage(filename):
    img1 = image.load_img(filename,target_size=img_size)
    plt.imshow(img1)
    Y = image.img_to_array(img1)
    X = np.expand_dims(Y,axis=0)
    X /= 255.0
    print(model.predict(X))
    val = argmax(model.predict(X),axis=1)
    print(val)
    if val == 1:
        plt.xlabel("Fire",fontsize=30)
    elif val == 0:
        plt.xlabel("Smoke",fontsize=30)
    else:
        plt.xlabel('Non Fire',fontsize=30)
```
